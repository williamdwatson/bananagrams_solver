```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
times = pd.read_excel("stats.xlsx", sheet_name="Time to solution")
densities = pd.read_excel("stats.xlsx", sheet_name="Board density")
```

```python
m, b = np.polyfit(np.repeat(times.columns, times.shape[0]), times.values.T.flatten(), 1)
m, b
```

```python
plt.scatter(np.repeat(times.columns, times.shape[0]), times.values.T.flatten())
plt.plot([10, 144], [m*10+b, m*144+b], '--k')
plt.xlabel("Number of letters in hand")
plt.ylabel("Time to solution (ms)")
plt.savefig("time_to_solution.png", dpi=300)
```

```python
m, b = np.polyfit(np.repeat(densities.columns, densities.shape[0]), densities.values.T.flatten(), 1)
m, b
```

```python
plt.scatter(np.repeat(densities.columns, densities.shape[0]), densities.values.T.flatten())
plt.plot([10, 144], [m*10+b, m*144+b], '--k')
plt.xlabel("Number of letters in hand")
plt.ylabel("Board density (letter/square)")
plt.savefig("board_density.png", dpi=300)
```

```python
plt.bar(range(len(times.columns)), times.mean(axis=0), yerr=times.std(axis=0), capsize=3)
plt.ylim(0)
plt.xticks(ticks=range(len(times.columns)), labels=times.columns)
plt.xlabel("Number of letters in hand")
plt.ylabel("Average time to solution (ms)")
plt.savefig("average_time_to_solution.png", dpi=300)
```

```python
plt.bar(range(len(densities.columns)), densities.mean(axis=0), yerr=densities.std(axis=0), capsize=3)
plt.ylim(0)
plt.xticks(ticks=range(len(densities.columns)), labels=densities.columns)
plt.xlabel("Number of letters in hand")
plt.ylabel("Average board density (letters/square)")
plt.savefig("average_board_density.png", dpi=300)
```

```python
for c in times.columns:
    print(c, times[c].mean(), times[c].std())
```

```python
for c in densities.columns:
    print(c, densities[c].mean(), densities[c].std())
```
